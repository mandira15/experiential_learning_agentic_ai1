```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Churn_Modelling (1) - Churn_Modelling (1).csv", encoding='latin')
df.head()
```

```python
numeric_df = df.select_dtypes(include=np.number)
cor = numeric_df.corr()
```

```python
top_cor = cor.index
plt.figure(figsize=(20,20))
sns.heatmap(df[top_cor].corr(),annot = True)
```

```python
x = df.iloc[: , 3:13 ]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
geo = pd.get_dummies(x["Geography"] ,drop_first=True)
geo
```

```python
gender = pd.get_dummies(x["Gender"] ,drop_first=True)
gender
```

```python
x=x.drop(["Geography" , "Gender"],axis = "columns")
```

```python
x.head(20)
```

```python
x = pd.concat([x,geo , gender] , axis="columns")
x.head(2)
```

```python
params ={
    "learning_rate":[0.05,0.10,0.15,0.20,0.25,0.30],
    "max_depth":[3,4,5,6,7,8,9,10,12],
    "min_child_weight":[1,3,5,7],
    "gamma":[0.0,0.1,0.2,0.3,0.4],

}
```

```python
from xgboost import XGBClassifier
mod = XGBClassifier()
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
grid = GridSearchCV(estimator = mod,param_grid=params,n_jobs=1,cv=3)
```

```python
grid.fit(x,y)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
m1 = XGBClassifier()
m1.fit(x,y)
```

```python
sc = cross_val_score(m1,x,y,cv=5)
```
